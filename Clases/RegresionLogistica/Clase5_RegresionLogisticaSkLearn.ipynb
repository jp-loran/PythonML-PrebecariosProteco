# Regresi√≥n Logistica
### Implementada con scikitLearn


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
iris= datasets.load_iris()
list(iris.keys())
iris
```

```python
X = iris['data'][:,3:]
Y = (iris['target']==2).astype(np.int)
```

```python
from sklearn.linear_model import LogisticRegression

modeloDeRegresionLogistica = LogisticRegression()
modeloDeRegresionLogistica.fit(X,Y)
```

```python
equis = np.linspace(0,3,1000).reshape(-1, 1)
lle = modeloDeRegresionLogistica.predict_proba(equis)
plt.plot(equis, lle[:,1], 'g-',label='Iris Virginica')
plt.plot(equis, lle[:,0], 'b--', label='No Iris Virginica')

plt.show()
```
